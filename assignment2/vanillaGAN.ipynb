```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import numpy as np

lr = 3e-4
epochs = 100
batch_size = 64

# MNIST
transform = transforms.ToTensor()
train_dataset = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

class Generator(nn.Module):
    def __init__(self):
        super(Generator, self).__init__()
        self.model = nn.Sequential(
            nn.Linear(100, 128),
            nn.LeakyReLU(0.2),
            nn.Linear(128, 256),
            nn.LeakyReLU(0.2),
            nn.Linear(256, 512),
            nn.LeakyReLU(0.2),
            nn.Linear(512, 1024),
            nn.LeakyReLU(0.2),
            nn.Linear(1024, 28*28),
            nn.Tanh()
        )

    def forward(self, x):
        return self.model(x)

class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()
        self.model = nn.Sequential(
            nn.Linear(28*28, 1024),
            nn.LeakyReLU(0.2),
            nn.Linear(1024, 512),
            nn.LeakyReLU(0.2),
            nn.Linear(512, 256),
            nn.LeakyReLU(0.2),
            nn.Linear(256, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        x = x.view(x.size(0), -1)
        return self.model(x)
```

```python
criterion = nn.BCELoss()
G = Generator()
D = Discriminator()
G_optimizer = optim.Adam(G.parameters(), lr=lr)
D_optimizer = optim.Adam(D.parameters(), lr=lr)

for epoch in range(epochs):
    for batch_idx, (x, _) in enumerate(train_loader):
        valid = torch.ones(x.size(0), 1)
        fake = torch.zeros(x.size(0), 1)

        real_imgs = x

        # Train Generator
        G_optimizer.zero_grad()
        z = torch.randn(x.size(0), 100)
        gen_imgs = G(z)
        G_loss = criterion(D(gen_imgs), valid)
        G_loss.backward()
        G_optimizer.step()

        # Train Discriminator
        D_optimizer.zero_grad()
        real_loss = criterion(D(real_imgs), valid)
        fake_loss = criterion(D(gen_imgs.detach()), fake)
        D_loss = (real_loss + fake_loss) / 2
        D_loss.backward()
        D_optimizer.step()

        if batch_idx % 100 == 0:
            print('Epoch: {}, Batch: {}, D_loss: {:.4f}, G_loss: {:.4f}'.format(epoch, batch_idx, D_loss.item(), G_loss.item()))

    if epoch % 10 == 0:
        z = torch.randn(1, 100)
        gen_imgs = G(z)
        gen_imgs = gen_imgs.view(28, 28).data.numpy()
        plt.imshow(gen_imgs, cmap='gray')
        plt.show()
        plt.close()

        
```

```python
import torch.autograd as autograd

LAMBDA = 10
n_critic = 5
lr = 1e-4

class Generator(nn.Module):
    def __init__(self):
        super(Generator, self).__init__()
        self.model = nn.Sequential(
            nn.Linear(100, 128),
            nn.LeakyReLU(0.2),
            nn.Linear(128, 256),
            nn.LeakyReLU(0.2),
            nn.Linear(256, 512),
            nn.LeakyReLU(0.2),
            nn.Linear(512, 1024),
            nn.LeakyReLU(0.2),
            nn.Linear(1024, 28*28),
            nn.Tanh()
        )

    def forward(self, x):
        return self.model(x)

class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()
        self.model = nn.Sequential(
            nn.Linear(28*28, 1024),
            nn.LeakyReLU(0.2),
            nn.Linear(1024, 512),
            nn.LeakyReLU(0.2),
            nn.Linear(512, 256),
            nn.LeakyReLU(0.2),
            nn.Linear(256, 1),
        )

    def forward(self, x):
        x = x.view(x.size(0), -1)
        return self.model(x)

def compute_gradient_penalty(D, real_samples, fake_samples):
    # Random weight term for interpolation between real and fake samples
    alpha = torch.rand((real_samples.size(0), 1), device=real_samples.device)
    alpha = alpha.expand_as(real_samples)
    
    # Get random interpolation between real and fake samples
    interpolates = (alpha * real_samples + ((1 - alpha) * fake_samples)).requires_grad_(True)
    d_interpolates = D(interpolates)
    fake = torch.ones(d_interpolates.size(), requires_grad=False, device=real_samples.device)

    # Get gradient w.r.t. interpolates
    gradients = autograd.grad(
        outputs=d_interpolates,
        inputs=interpolates,
        grad_outputs=fake,
        create_graph=True,
        retain_graph=True,
        only_inputs=True,
    )[0]

    gradients = gradients.view(gradients.size(0), -1)
    gradient_penalty = ((gradients.norm(2, dim=1) - 1) ** 2).mean()
    return gradient_penalty

G = Generator()
D = Discriminator()
G_optimizer = optim.Adam(G.parameters(), lr=lr, betas=(0.0, 0.9))
D_optimizer = optim.Adam(D.parameters(), lr=lr, betas=(0.0, 0.9))

for epoch in range(epochs):
    for batch_idx, (real_imgs, _) in enumerate(train_loader):
        
        real_imgs = real_imgs.view(real_imgs.size(0), -1)

        D_optimizer.zero_grad()
        
        z = torch.randn(real_imgs.size(0), 100)

        gen_imgs = G(z)

        real_loss = -torch.mean(D(real_imgs))
        fake_loss = torch.mean(D(gen_imgs.detach()))
        gradient_penalty = compute_gradient_penalty(D, real_imgs.data, gen_imgs.data)
        D_loss = real_loss + fake_loss + LAMBDA * gradient_penalty
        
        D_loss.backward()
        D_optimizer.step()

        if batch_idx % n_critic == 0:
            G_optimizer.zero_grad()

            gen_imgs = G(z)
            G_loss = -torch.mean(D(gen_imgs))
            
            G_loss.backward()
            G_optimizer.step()

        if batch_idx % 100 == 0:
            print('Epoch: {}, Batch: {}, D_loss: {:.4f}, G_loss: {:.4f}'.format(
                epoch, batch_idx, D_loss.item(), G_loss.item()
            ))

    if epoch % 10 == 0:
        z = torch.randn(1, 100)
        gen_imgs = G(z)
        gen_imgs = gen_imgs.view(28, 28).data.numpy()
        plt.imshow(gen_imgs, cmap='gray')
        plt.show()
        plt.close()
```

```python
for param in G.parameters():
    param.requires_grad = False

def optimize_z(G, x, iterations=1000, lr=0.01):
    z = torch.randn(1, 100, requires_grad=True)
    optimizer = optim.Adam([z], lr=lr)

    for i in range(iterations):
        optimizer.zero_grad()
        x_hat = G(z)
        loss = ((x_hat - x)**2).mean()
        loss.backward()
        optimizer.step()

    return z.detach()

test_dataset = datasets.MNIST(root='./data', train=False, download=True, transform=transform)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=1, shuffle=True)

for i, (x, _) in enumerate(test_loader):
    if i >= 5:
        break
    x = x.view(x.size(0), -1)
    z_star = optimize_z(G, x)
    print(f"Optimized z for image {i+1}: {z_star}")

    z_star2 = optimize_z(G, x)
    print(f"Different initialization leads to: {z_star2}")
    print(f"Do the embeddings match? {torch.allclose(z_star, z_star2, atol=1e-4)}")

    with torch.no_grad():
        reconstructed_img = G(z_star).view(28, 28).cpu().numpy()
        original_img = x.view(28, 28).cpu().numpy()
        
        plt.figure(figsize=(4, 2))
        plt.subplot(1, 2, 1)
        plt.title("Original")
        plt.imshow(original_img, cmap='gray')
        plt.subplot(1, 2, 2)
        plt.title("Reconstructed")
        plt.imshow(reconstructed_img, cmap='gray')
        plt.show()
```
